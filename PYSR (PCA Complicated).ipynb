```python
from pysr import PySRRegressor
import numpy as np
```

```python
df = np.load('./nac_complicated.npz')
X = df['arr_0']
X_hat = df['arr_1']
y = df['arr_2']
```

```python
X.shape
```

```python
X_hat.shape
```

## Symbolic Regression for Original Data

```python
model_1 = PySRRegressor(
    niterations=5,
    binary_operators=["+", "*", "-"],
    unary_operators=["sqrt", "cos", "sin"],
    model_selection="best",
    loss="loss(x,y) = (x-y)^2",
)
```

```python
model_1.fit(X, y) # Original
```

```python
model_1.equations
```

## Symbolic Regression for NAC filtered Data

```python
model_2 = PySRRegressor(
    niterations=5,
    binary_operators=["+", "*", "-"],
    unary_operators=["sqrt", "sin", "cos"],
    model_selection="best",
    loss="loss(x,y) = (x-y)^2",
)
```

```python
model_2.fit(X_hat, y)
```

```python
model_2.equations
```
