```python
var('x y')
```

```python
sigmoid(x) = 1 / (1 + exp(-x))
```

```python
f(x, y) = tanh(x) * sigmoid(y)
```

```python
plot3d(f, (x, -10, 10), (y, -10, 10))
```

```python
plot(tanh(x), (x, -10, 10)) + plot(diff(tanh(x), x), (x, -10, 10))
```

```python
plot(sigmoid(y), (y, -10, 10)) + plot(diff(sigmoid(y), y), (y, -10, 10))
```

```python
g(x, y) = tanh(x)**2 * sigmoid(y)
plot3d(g, (x, -10, 10), (y, -10, 10))
```

```python
h(x, y) = sigmoid(x) * sigmoid(y)
plot3d(h, (x, -10, 10), (y, -10, 10))
```

```python
s(x, y) = sigmoid(y)
plot3d(s, (x, -10, 10), (y, -10, 10))
```

```python
jacobian(f, (x, y))(-10.0, 10.0)
```

```python
jacobian(g, (x, y))(-10.0, 10.0)
```

```python
jacobian(h, (x, y))(-10.0, 10.0)
```

```python
jacobian(s, (x, y))(-10.0, 10.0)
```
